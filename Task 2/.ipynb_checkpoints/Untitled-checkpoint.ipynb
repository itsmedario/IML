```python
import numpy as np
```

```python
# Read the data and remove the first Row
train_data = np.genfromtxt('train_features.csv', delimiter = ',') 
train_data = train_data[1:]
```

```python
# Make an array of data arrays X = [[x00,x01,...,x09],...]
X_all = np.array([row[2:] for row in train_data])
```

```python
# Make group of 12 hour vectors
X_subgroups = np.array_split(X_all, len(X_all)/12)
X_subgroups[0:3]
```

```python
X_avg = []
for sub in X_subgroups[0:3]:
    X_curr = [[0,0]] * len(sub[0])
    for list in sub:
        for i, item in enumerate(list, 0):
            if item != 'nan':
                X_curr[i][0] += item
                X_curr[i][1] += 1
    print(X_curr)
    X_avg.append([item[0]/item[1] if item[1] != 0 else 'nan' for item in X_curr])
X_avg[0:3]
```
